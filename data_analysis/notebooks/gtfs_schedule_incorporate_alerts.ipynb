```python
# Unwrap - get start and end
# TODO - prod code
alerts_deduped_df['alert_start'] = pd.to_datetime(alerts_deduped_df['alert.activePeriod'].apply(lambda x: x[0]['start']),unit='s', utc=True)
alerts_deduped_df['alert_end'] = pd.to_datetime(alerts_deduped_df['alert.activePeriod'].apply(lambda x: x[0]['end']),unit='s', utc=True)
# alerts_deduped_df['alert_end'] = pd.to_datetime(alerts_deduped_df['alert_end'])
# Advance notice of alert: Timedelta between header.timestamp in deduped (sorted) and alert_start
alerts_deduped_df['header.timestamp'] = alerts_deduped_df['header.timestamp'].dt.tz_localize('America/Denver') # Only run once
alerts_deduped_df['advance_notice'] = alerts_deduped_df['alert_start'] - alerts_deduped_df['header.timestamp']
alerts_deduped_df['advance_notice'].sort_values() # Handful of negatives. Long-running before data collection? # Not Code
# alerts_deduped_df['advance_notice'].plot(kind='hist')
alerts_deduped_df.iloc[4344] # Not Code
alerts_deduped_df.loc[alerts_deduped_df['advance_notice'] >= pd.Timedelta(-1,'s'),'advance_notice'].dt.total_seconds().plot.hist() # # Not Code

# Fill in end date with max(header.timestamp) when end is empty & max(header.timestamp) for those rows is *not* the max across alerts_df (indicating that it's still ongoing)
no_end_list = alerts_deduped_df.loc[alerts_deduped_df['alert_end'].isna()]['id'].tolist()
end_fill_df = alerts_df.loc[alerts_df['id'].isin(no_end_list)].groupby('id')['header.timestamp'].max()
print(end_fill_df.shape) # Not Code
end_fill_df = end_fill_df.loc[end_fill_df < alerts_df['header.timestamp'].max()]
end_fill_df.name = 'alert_end' # Rename to successfully use update later
print(end_fill_df.shape) # Not Code - 19 still ongoing as of May 1-ish
alerts_deduped_df2 = alerts_deduped_df.set_index('id')
alerts_deduped_df2.update(end_fill_df.dt.tz_localize('America/Denver'))
alerts_deduped_df2['alert_end'] = pd.to_datetime(alerts_deduped_df2['alert_end'], utc=True) # Have to re-cast type after update

print(alerts_deduped_df2['alert_end'].isna().value_counts()) # Not Code

alerts_deduped_df = alerts_deduped_df2.reset_index()
# alerts_deduped_df2.index
```

```python
# TODO Code to extract affected stops + trips
alerts_deduped_df.head()
# alerts_deduped_df['alert.informedEntity'].iloc[50][0] # For a given row, list of dicts with routeId:/stopId:/trip:{tripId:<string of trip id>)
alerts_deduped_df['affected_stops_list'] = alerts_deduped_df['alert.informedEntity'].apply(lambda x: [d['stopId'] for d in x]) # Stop Id's are all lists
alerts_deduped_df['affected_routes_list'] = alerts_deduped_df['alert.informedEntity'].apply(lambda x: [d['routeId'] for d in x])
alerts_deduped_df['affected_trips_list'] = alerts_deduped_df['alert.informedEntity'].apply(lambda x: [d['trip']['tripId'] for d in x if d['trip'] is not None] ) # Note, all ints
# alerts_deduped_df['alert.effect'].value_counts()
# ['NO_SERVICE', 'DETOUR', 'MODIFIED_SERVICE'] # Use these categories in alert.effect to flag reduced service 
```

```python
# TODO Code to apply these to the trips_scheduled / stops_scheduled datasets
# merge between stop and end
 #Can also try pandasql, but this supposedly has better performance

# Create dataframes with start time/stop time/affected (route/trip/stop)
# One dataframe per kind
# Drop duplicates to cover for cases where the alerts repeat the same info

alerts_deduped_df['affected_stops_list'].explode() # index is numeric index, not alert id.
alerts_cols = ['id','alert_start', 'alert_end','advance_notice','alert.cause', 'alert.effect', 'alert.headerText.translation', 'alert.descriptionText.translation']
alerts_duplicate_cols = ['id','alert_start', 'alert_end','advance_notice','alert.cause', 'alert.effect']
affected_stops_df = alerts_deduped_df[alerts_cols].merge(alerts_deduped_df['affected_stops_list'].explode(), left_index = True, right_index = True)
affected_stops_df['affected_stops_list'] = affected_stops_df['affected_stops_list'].astype(str)
affected_stops_df = affected_stops_df.drop_duplicates(alerts_duplicate_cols)

affected_trips_df = alerts_deduped_df[alerts_cols].merge(alerts_deduped_df['affected_trips_list'].explode(), left_index = True, right_index = True)
affected_trips_df['affected_trips_list'] = affected_trips_df['affected_trips_list'].astype(str)
affected_stops_df = affected_stops_df.drop_duplicates(alerts_duplicate_cols)

affected_routes_df = alerts_deduped_df[alerts_cols].merge(alerts_deduped_df['affected_routes_list'].explode(), left_index = True, right_index = True)
affected_routes_df['affected_routes_list'] = affected_routes_df['affected_routes_list'].astype(str)
affected_routes_df = affected_routes_df.drop_duplicates(alerts_duplicate_cols)
```

```python
# relational version:
# 10 seconds to do a big merge and filter on start and end times!
# Prod code:
con = duckdb.connect()
con.execute("SET GLOBAL pandas_analyze_sample=100000") # Necessary to avoid typing errors. Or specify manually.
# trips_scheduled_rel = con.from_df(trips_scheduled).set_alias('trips_scheduled_rel')
stops_scheduled_rel = con.from_df(stops_scheduled.reset_index()).set_alias('stops_scheduled_rel')
# stops_scheduled_rel = con.from_df(stops_scheduled[:10000].reset_index()).set_alias('stops_scheduled_rel') # Sample for development

# stops_scheduled_rel = con.from_df(stops_scheduled.reset_index()).set_alias('stops_scheduled_rel')
affected_stops_rel = con.from_df(affected_stops_df).set_alias('affected_stops_rel')
affected_trips_rel = con.from_df(affected_trips_df).set_alias('affected_trips_rel')
affected_routes_rel = con.from_df(affected_routes_df).set_alias('affected_routes_rel')

# Get affected stops/trips/routes

# Only get cols necessary 


# affected_stops_rel = affected_stops_rel[stops_affected_keep_cols]
# affected_trips_rel = affected_trips_rel[affected_trips_rel]
# affected_routes_rel = affected_routes_rel[affected_routes_rel]

# stops_scheduled_rel.join(affected_stops_rel,'stops_scheduled_rel.stop_id = affected_stops_rel.affected_stops_list',how = 'left').filter('affected_stops_rel.alert_start < stops_scheduled_rel.departure_datetime').filter('affected_stops_rel.alert_end > stops_scheduled_rel.departure_datetime').show() #
# print(stops_scheduled_affected_stops_df.shape)
# print(stops_scheduled_affected_stops_df.head() )

stops_scheduled_affected_stops_rel = (stops_scheduled_rel
                                     .join(affected_stops_rel,'stops_scheduled_rel.stop_id = affected_stops_rel.affected_stops_list',how = 'inner')
                                     .filter('affected_stops_rel.alert_start < stops_scheduled_rel.departure_datetime')
                                     .filter('affected_stops_rel.alert_end > stops_scheduled_rel.departure_datetime')
                                     .order('index,advance_notice')
                                     .distinct())
                                    #  .df()) # Might need to dedupe again later on index
# .df()
stops_scheduled_affected_trips_rel = (stops_scheduled_rel
                                     .join(affected_trips_rel,'stops_scheduled_rel.trip_id = affected_trips_rel.affected_trips_list',how = 'inner')
                                     .filter('affected_trips_rel.alert_start < stops_scheduled_rel.departure_datetime')
                                     .filter('affected_trips_rel.alert_end > stops_scheduled_rel.departure_datetime')
                                    .order('index,advance_notice')
                                     .distinct())
                                    #  .df()) # Might need to dedupe again later on index

                                    # .df())
stops_scheduled_affected_routes_rel = (stops_scheduled_rel
                                      .join(affected_routes_rel,'stops_scheduled_rel.route_id = affected_routes_rel.affected_routes_list',how = 'inner')
                                      .filter('affected_routes_rel.alert_start < stops_scheduled_rel.departure_datetime')
                                      .filter('affected_routes_rel.alert_end > stops_scheduled_rel.departure_datetime')
                                      .order('index,advance_notice')
                                      .distinct())
                                    #   .df()) # Might need to dedupe again later on index

                                    #   .df())
# stops_scheduled_df = stops_scheduled_rel.df() # Do I need this?

# Actually, I can optimize this too! # TODO - Bring this and merge into relation before calling .df()
# print(stops_scheduled_affected_stops_df.shape)
# print(stops_scheduled_affected_trips_df.shape)
# print(stops_scheduled_affected_routes_df.shape)
# stops_scheduled_affected_stops_df = stops_scheduled_affected_stops_df.sort_values('advance_notice', ascending = False).drop_duplicates('index')
# stops_scheduled_affected_trips_df = stops_scheduled_affected_trips_df.sort_values('advance_notice', ascending = False).drop_duplicates('index')
# stops_scheduled_affected_routes_df = stops_scheduled_affected_routes_df.sort_values('advance_notice', ascending = False).drop_duplicates('index')
# print(stops_scheduled_affected_stops_df.shape)
# print(stops_scheduled_affected_trips_df.shape)
# print(stops_scheduled_affected_routes_df.shape)

stops_scheduled_affected_stops_rel.to_table('stops_scheduled_affected_stops_tbl')
stops_scheduled_affected_trips_rel.to_table('stops_scheduled_affected_trips_tbl')
stops_scheduled_affected_routes_rel.to_table('stops_scheduled_affected_routes_tbl')
# alert.headerText.translation', 'alert.descriptionText.translation
# = stops_scheduled_affected_stops_rel.query("select 'index', 'alert_start', 'alert_end', 'advance_notice', 'alert.cause', 'alert.effect', 'affected_stops_list' as ")
stops_subset_rel = con.sql('select index, alert_start_stops, alert_end_stops,advance_notice_stops,alert_cause_stops,alert_effect_stops\
                      alert_headerText_translation_stops,alert_descriptionText_translation_stops\
                           from (select  index, \
                      alert_start as alert_start_stops, \
                      alert_end as alert_end_stops,  \
                      advance_notice as advance_notice_stops, \
                      "alert.cause" as alert_cause_stops, \
                      "alert.effect" as alert_effect_stops, \
                      affected_stops_list as affected_stops_list_stops, \
                        "alert.headerText.translation" as alert_headerText_translation_stops,\
                          "alert.descriptionText.translation" as alert_descriptionText_translation_stops,\
                        rank() over (partition by index order by advance_notice) as rank \
                      from stops_scheduled_affected_stops_tbl) a \
                        where a.rank=1').set_alias('stops_subset_rel')
trips_subset_rel = con.sql('select index, alert_start_trips, alert_end_trips,advance_notice_trips,alert_cause_trips,alert_effect_trips\
                      alert_headerText_translation_trips,alert_descriptionText_translation_trips\
                           from (select  index, \
                      alert_start as alert_start_trips, \
                      alert_end as alert_end_trips,  \
                      advance_notice as advance_notice_trips, \
                      "alert.cause" as alert_cause_trips, \
                      "alert.effect" as alert_effect_trips, \
                      affected_trips_list as affected_trips_list_trips, \
                          "alert.headerText.translation" as alert_headerText_translation_trips,\
                          "alert.descriptionText.translation" as alert_descriptionText_translation_trips,\
                        rank() over (partition by index order by advance_notice) as rank \
                      from stops_scheduled_affected_trips_tbl) a \
                        where a.rank=1').set_alias('trips_subset_rel')


routes_subset_rel = con.sql('select index, alert_start_routes, alert_end_routes,advance_notice_routes,alert_cause_routes,alert_effect_routes\
                      alert_headerText_translation_routes,alert_descriptionText_translation_routes\
                           from (select  index, \
                      alert_start as alert_start_routes, \
                      alert_end as alert_end_routes,  \
                      advance_notice as advance_notice_routes, \
                      "alert.cause" as alert_cause_routes, \
                      "alert.effect" as alert_effect_routes, \
                      affected_routes_list as affected_list_routes, \
                                "alert.headerText.translation" as alert_headerText_translation_routes,\
                          "alert.descriptionText.translation" as alert_descriptionText_translation_routes,\
                        rank() over (partition by index order by advance_notice) as rank \
                      from stops_scheduled_affected_routes_tbl) a \
                        where a.rank=1').set_alias('routes_subset_rel')

# stops_subset_rel.to_table('stops_subset_rel')
# trips_subset_rel.to_table('trips_subset_rel')
# routes_subset_rel.to_table('routes_subset_rel')


# Stick the whole thing in a parquet
grand_finale_df = (stops_scheduled_rel
                   .join(stops_subset_rel, 'stops_scheduled_rel.index = stops_subset_rel.index',how = 'left')
                   .join(trips_subset_rel, 'stops_scheduled_rel.index = trips_subset_rel.index',how = 'left')
                   .join(routes_subset_rel, 'stops_scheduled_rel.index = routes_subset_rel.index',how = 'left')
                   ).df()
grand_finale_df.to_parquet(f's3://rtd-ghost-buses-{BUCKET_TYPE}/schedules/scheduled_trips_with_alerts_df.parquet')


# TODO Merge deduped stops_scheduled_affected_stops_df/et cetera back to stops_scheduled on 'index' after filtering out 
# stops_affected_keep_cols = ['index', 'alert_start', 'alert_end', 'advance_notice', 'alert.cause', 'alert.effect', 'affected_stops_list']
# trips_affected_keep_cols = ['index', 'alert_start', 'alert_end', 'advance_notice', 'alert.cause', 'alert.effect', 'affected_trips_list']
# routs_affected_keep_cols = ['index', 'alert_start', 'alert_end', 'advance_notice', 'alert.cause', 'alert.effect', 'affected_routes_list']

# stops_scheduled_affected_stops_df = stops_scheduled_affected_stops_df[stops_affected_keep_cols]
# stops_scheduled_affected_trips_df = stops_scheduled_affected_trips_df[trips_affected_keep_cols]
# stops_scheduled_affected_routes_df = stops_scheduled_affected_routes_df[routs_affected_keep_cols]

# stops_scheduled_rel.join(stops_scheduled_affected_stops_df)

# TODO Write out to parquet?


# output_df = transformed_rel.from_df()
con.close()
```
